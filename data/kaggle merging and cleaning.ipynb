```python
import pandas as pd
import os

# Path to your dataset folder
data_path = "/kaggle/input/behavioral-risk-factor-surveillance-system/"

# List all CSV files in the folder
files = [f for f in os.listdir(data_path) if f.endswith(".csv")]
print("Files found:", files)

# Load and concatenate all years
df_list = []
for f in sorted(files):  # sorted to load 2011 -> 2015
    print("Loading:", f)
    df_year = pd.read_csv(os.path.join(data_path, f))
    df_year['YEAR'] = int(f[:4])  # add YEAR column for reference
    df_list.append(df_year)

brfss_data = pd.concat(df_list, ignore_index=True)
print("Merged shape:", brfss_data.shape)
```

```python
# Step 2: Keep only important columns
cols_to_keep = [
    "AGE", "SEX", "MARITAL", "EDUCA", "EMPLOY", "INCOME2",
    "GENHLTH", "PHYSHLTH", "MENTHLTH", "POORHLTH",
    "BPHIGH4", "BPMEDS", "BLOODCHO", "CHOLCHK",
    "CVDINFR4", "CVDCRHD4", "CVDSTRK3",
    "DIABETE3", "PREDIAB1", "BLDSUGAR", "INSULIN", "DIABAGE2",
    "WEIGHT2", "HEIGHT3", "_BMI5", "_BMI5CAT",
    "SMOKE100", "SMOKDAY2", "ALCDAY5", "EXERANY2",
    "HLTHPLN1", "PERSDOC2", "MEDCOST", "CHECKUP1"
]

df = brfss_data[cols_to_keep]  # <- use brfss_data, not merged_data

print("Shape after keeping only selected columns:", df.shape)
```

```python
# Save to CSV
df.to_csv("brfss_cleaned.csv", index=False)
print("File saved as brfss_cleaned.csv")
```
